```python
from sqlalchemy import create_engine
import pymysql

db_connection_str = 'mysql+pymysql://root:adepu321@localhost/CVD_GEORGIA_RATES'
db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT * FROM health_risk', con=db_connection)

df
```

```python
physical_inactivity = df[["physical_inactivity"]]
print(physical_inactivity)

obeisity = df[["obeisity"]]
print(obeisity)

diabetes = df[["diabetes"]]
print(diabetes)


df["risk_value_pi"] = ""
df["risk_value_obs"] = ""
df["risk_value_diab"] = ""

print (df)
```

```python
df.loc[df['physical_inactivity'] > 26.48, 'risk_value_pi'] = .005
df.loc[df['physical_inactivity'] < 26.48, 'risk_value_pi'] = .01

# mean physical activity = 26.48
# the higher percentage of populations that is inactive, the higher the CVD mortality rate

df.loc[df['obeisity'] > 31.71, 'risk_value_obs'] = .01
df.loc[df['obeisity'] < 31.71, 'risk_value_obs'] = .005

# mean obeisity = 31.71

df.loc[df['diabetes'] > 11.51, 'risk_value_diab'] = .01
df.loc[df['diabetes'] < 11.51, 'risk_value_diab'] = .005

# mean diabetes = 11.51

df
```
