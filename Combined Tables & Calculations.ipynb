```python
from sqlalchemy import create_engine
import pymysql

db_connection_str = 'mysql+pymysql://root:adepu321@localhost/CVD_GEORGIA_RATES'
db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT * FROM georgia_death_rates', con=db_connection)

death_rates = df[["death_rates"]]
print(death_rates)

# df = pd.DataFrame(death_rates,columns=['death_rates'])
# print (df)

df["death_risk"] = ""

df.loc[df['death_rates'] > 500, 'death_risk'] = .01
df.loc[df['death_rates'] < 500, 'death_risk'] = .005

print (df)
```

```python
df1 = df = pd.read_sql('SELECT * FROM social_environment', con=db_connection)

median_hi = df[["median_household_income"]]
print(median_hi)

poverty_perc = df[["poverty_percentage"]]
print(poverty_perc)

# df = pd.DataFrame(death_rates,columns=['death_rates'])
# print (df)


df["risk_value_hi"] = ""
df["risk_value_pp"] = ""

df.loc[df['median_household_income'] > 42610, 'risk_value_hi'] = .005
df.loc[df['median_household_income'] < 42610, 'risk_value_hi'] = .01

#mean median household income = 42610 for Georgia according to calculations

df.loc[df['poverty_percentage'] > 21.91, 'risk_value_pp'] = .01
df.loc[df['poverty_percentage'] < 21.91, 'risk_value_pp'] = .005

#mean poverty percentage = 21.91 for Georgia according to calculations

df
```

```python
df2 = df = pd.read_sql('SELECT * FROM health_risk', con=db_connection)

physical_inactivity = df[["physical_inactivity"]]
print(physical_inactivity)

obeisity = df[["obeisity"]]
print(obeisity)

diabetes = df[["diabetes"]]
print(diabetes)


df["risk_value_pi"] = ""
df["risk_value_obs"] = ""
df["risk_value_diab"] = ""

df.loc[df['physical_inactivity'] > 26.48, 'risk_value_pi'] = .005
df.loc[df['physical_inactivity'] < 26.48, 'risk_value_pi'] = .01

# mean physical activity = 26.48
# the higher percentage of populations that is inactive, the higher the CVD mortality rate

df.loc[df['obeisity'] > 31.71, 'risk_value_obs'] = .01
df.loc[df['obeisity'] < 31.71, 'risk_value_obs'] = .005

# mean obeisity = 31.71

df.loc[df['diabetes'] > 11.51, 'risk_value_diab'] = .01
df.loc[df['diabetes'] < 11.51, 'risk_value_diab'] = .005

# mean diabetes = 11.51

print (df)

```

```python
df3 = df = pd.read_sql('SELECT * FROM physical_environment', con=db_connection)

air_quality = df[["air_quality"]]
print(air_quality)

park_half_mile = df[["park_in_half_mile"]]
print(park_half_mile)

df["risk_value_aq"] = ""
df["risk_value_phm"] = ""

df.loc[df['air_quality'] > 10.74, 'risk_value_aq'] = .01
df.loc[df['air_quality'] < 10.74, 'risk_value_aq'] = .005

# mean air quality (concentrations of PM2.5) = 10.74
# the higher the PM2.5 concentration for air quality, the higher the CVD mortality rate 

df.loc[df['park_in_half_mile'] > 7.59, 'risk_value_phm'] = .005
df.loc[df['park_in_half_mile'] < 7.59, 'risk_value_phm'] = .01

# mean number of parks in half mile = 7.59 --> Round down to 7 

print(df3)
```
